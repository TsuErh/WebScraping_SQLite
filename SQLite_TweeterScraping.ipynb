```python
# Connect to database
conn = sqlite3.connect('WebScraping.db') 
cursor = conn.cursor()
```

```python
cursor.execute('DROP TABLE User;')
cursor.execute('DROP TABLE Tweets;')
```

```python
# Create table "user"
UserTable = """
    CREATE TABLE User (
        ID NUMBER(20),
        Name VARCHAR2(40),
        ScreenName VARCHAR2(40),
        Description VARCHAR2(400),
        FriendsCount NUMBER(10),
        
        CONSTRAINT UserPK
        PRIMARY KEY(ID)
    );
"""
```

```python
# Create Table "Tweets"
TweetsTable = """
    CREATE TABLE Tweets(
        ID NUMBER(20),
        created_at VARCHAR2(50),
        id_str NUMBER(25),
        text VARCHAR2(50),
        source VARCHAR(80),
        in_reply_to_user_id NUMBER(25),
        in_reply_to_screen_name VARCHAR2(25),
        in_reply_to_status_id NUMBER(25),
        retweet_count NUMBER(5),
        contributors VARCHAR(25),
        
        CONSTRAINT Tweets_PK
        PRIMARY KEY(id_str),
        
        CONSTRAINT Tweets_FK
        FOREIGN KEY(ID) REFERENCES User
    );
"""
```

```python
cursor.execute(UserTable)
```

```python
cursor.execute(TweetsTable)
```

```python
# Insert values into table after web scraping
All = urllib.request.urlopen('https://dbgroup.cdm.depaul.edu/DSC450/Module7.txt')
Error = open('Module7_errors.txt', 'a')
for tweet in All:
    try:
        tdict = json.loads(tweet.decode('utf8'))
        
        values = []
        ID = tdict['id']
        created_at = tdict['created_at']
        id_str = tdict['id_str']
        text = tdict['text']
        source = tdict['source']
        in_reply_to_user_id = tdict['in_reply_to_user_id']
        in_reply_to_screen_name = tdict['in_reply_to_screen_name']
        in_reply_to_status_id = tdict['in_reply_to_status_id']
        retweet_count = tdict['retweet_count']
        contributors = tdict['contributors']
        values.extend((ID,created_at, id_str, text, source, in_reply_to_user_id, in_reply_to_screen_name, in_reply_to_status_id, retweet_count, contributors))
        #print(values)
        cursor.execute("INSERT OR IGNORE INTO Tweets VALUES (?,?,?,?,?,?,?,?,?,?);", values)
        
        UsrVal = []
        Name = tdict['user']['name']
        ScreenName = tdict['user']['screen_name']
        Description = tdict['user']['description']
        FriendsCount = tdict['user']['friends_count']
        UsrVal.extend((ID, Name, ScreenName, Description, FriendsCount))
        #print(UsrVal)
        cursor.execute("INSERT OR IGNORE INTO User VALUES (?,?,?,?,?);", UsrVal)
    except ValueError:
        Error.write(str(tweet))
```
